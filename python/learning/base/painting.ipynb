```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

```python
plt.rcParams['font.sans-serif']=['SimHei'] # 设置中文
plt.rcParams['axes.unicode_minus']=False # 设置负号
plt.rcParams['font.size'] = 12
```

```python
# 绘制正弦函数和余弦函数
# 大致流程是: 生成figure对象和axes对象 通过axes绘制数据(指定数据 确定标签 颜色 线宽)设置样式等
X = np.linspace(0, 2*np.pi, 100)
sin_y = np.sin(X)
cos_y = np.cos(X)
fig, ax = plt.subplots(figsize=(8,6))
# fig, (ax1,ax2) = plt.subplots(1,2,figsize=(8,6),sharey=True) 共享y轴
# fig = plt.figure() 
# ax1 = fig.add_subplot(2,1,1) 前两个参数表示图的行列数2行1列 第三个参数表示图的索引 从左到右 从上到下
# ax2 = fig.add_subplot(2,1,2)
ax.plot(X, sin_y, label='sin', color='b', linewidth=2)
ax.plot(X, cos_y, label='cos', color='r', linewidth=2)
ax.set_xticks(np.arange(0,2*np.pi+1,np.pi/2)) # 设置横轴刻度的位置 之所以要在第二个参数后加个1是因为左闭右开原则
ax.set_xticklabels([r'$0$', r'$\frac{\pi}{2}$', # 设置横轴刻度的标签 不然这里显示的是pi/2的小数形式
                    r'$\pi$', r'$\frac{3\pi}{2}$', 
                    r'$2\pi$'])
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_xlim(0, 2*np.pi)
ax.set_ylim(-1.5,1.5)
ax.legend()
# plt.subplots_adjust(wspace=0.5, hspace=0.5) 控制子图之间的间距 wspace控制横向间距
ax.set_aspect('equal') # 横纵轴采用相同的scale x的一格和y的一格长度上相同
ax.set_title('sin & cos')
plt.grid() # 添加背景格网
# plt.savefig('正弦_余弦函数曲线.svg', format='svg')
plt.show()
```

```python
# 只绘制一张图时 可以不创建axes对象 直接使用plt对象
x = np.linspace(-np.pi, np.pi, 100)
cos_y = np.cos(x)
sin_y = np.sin(x)
style = {
    'linewidth':3,
    'linestyle':':',
    'marker':'o',
    'color':'yellowgreen',
    'label':'y=sin(x)'
}
plt.plot(x, cos_y, label = 'y=cos(x)')
plt.plot(x, sin_y, **style)
plt.ylabel('y')
plt.xlabel('x')
plt.legend(loc = 'best')
plt.annotate(text='标记点',xy=(3,np.sin(3)),xytext=(2,-0.5),weight='bold',color='r',arrowprops=dict(arrowstyle='-|>',color='k'))
plt.show()
```

```python
# 绘制散点图
iris = load_iris()
sepal_length = iris.data[:,0]
sepal_width = iris.data[:,1]
target = iris.target

fig, ax = plt.subplots()
#横坐标 纵坐标 散点大小 颜色:单一颜色|颜色序列|数据序列 颜色映射 散点形状
plt.scatter(sepal_length, sepal_width, s=40, c=target, cmap='rainbow', marker='o')
plt.title('Iris sepal length vs width')
plt.xlabel('Sepal length (cm)')
plt.ylabel('Sepal width (cm)')

ax.set_xticks(np.arange(4,8+1))
ax.set_yticks(np.arange(1,5+1))
ax.axis('scaled')
ax.grid(linestyle='--', linewidth=0.25, color=[0.7,0.7,0.7])
ax.set_xbound(lower=4,upper=8)
ax.set_ybound(lower=1,upper=5)

plt.show()
```

```python
# 示例数据集
data = np.random.uniform(0,10,size=100)

# bins参数规定了直方图中柱子的数量 如果是整数则代表柱子的个数 是序列则代表柱子的边界
plt.hist(data, bins=10, align='left', color='blue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Frequency Histogram')

plt.show()
```
